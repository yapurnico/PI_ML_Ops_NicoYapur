```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import json_normalize
import ast
```

```python
movies_df = pd.read_csv("ETL_movies_main.csv")
```

```python
movies_df
```

```python
# Dropping Duplicates, just in case
movies_df = movies_df.drop_duplicates(keep="first")
```

```python
movies_df.shape
```

```python
# filtering needed columns
df_collection = movies_df.loc[:,["Movie_id","belongs_to_collection"]]
df_collection
```

```python
# droping NaN 
df_collection.dropna(axis=0,inplace=True)
```

```python
#chusmeo
df_collection
```

```python
# set id as index
df_collection.set_index('Movie_id',inplace=True)
```

```python
df_collection
```

```python
# transforming dictionaries to str
df_collection= df_collection['belongs_to_collection'].apply(ast.literal_eval) 
```

```python
# chusmeo
df_collection
```

```python
# set Movie_id as index
df_collection = df_collection.reset_index()
df_collection
```

```python
df = df_collection

# Convert dictionaries to columns
df = pd.concat([df, df['belongs_to_collection'].apply(pd.Series)], axis=1)
df.drop(columns=['belongs_to_collection'], inplace=True)

print(df)
```

```python
# dropping unnecesary columns
df_collection = df.iloc[:,[0,2,3,4]]
df_collection.rename(columns={"name":"collection_name","":""},inplace=True)
df_collection
```

```python
# df_collection to csv
```

```python
df_collection.to_csv("ETL_movies_collection.csv",index=False)
```
